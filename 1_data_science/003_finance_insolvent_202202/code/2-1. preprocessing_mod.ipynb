# outlier

## v1. 컬럼별 이상치 값들을 전, 후 회계년도 값으로 대체

```python
def outlier_replace_nan(df):
  col_list = df.drop(['회사명', '거래소코드', '회계년도', '산업코드', '소속코드', '상장일', '상장폐지일', '감사의견코드', '산업대분류'], axis=1).columns
  for each_col in tqdm(col_list):
    Q1 = np.quantile(df[each_col], 0.25)
    Q3 = np.quantile(df[each_col],0.75)

    minimum = Q1 - 1.5 * (Q3-Q1)
    maximum = Q3 + 1.5 * (Q3-Q1)

    # 각 컬럼별 이상치 값들을 다 결측치로 처리
    df.loc[(df[each_col] < minimum) | (df[each_col] > maximum), each_col] = np.nan
    return df

def outlier_fillna(df):
  code_list = df['거래소코드'].unique().tolist()

  list_of_df = [] 
  for each_code in code_list:
    tmp = df[df['거래소코드'] == each_code]
    tmp.fillna(method='ffill', inplace=True)
    tmp.fillna(method='bfill', inplace=True)
    list_of_df.append(tmp)

  new_df = pd.concat(list_of_df)

  return new_df

df_07_v1 = outlier_fillna(outlier_replace_nan(df_07))
```

## v2. 극단치를 제거

```python
def outlier_replace_nan_v2(df):
  col_list = df.drop(['회사명', '거래소코드', '회계년도', '산업코드', '소속코드', '상장일', '상장폐지일', '감사의견코드', '산업대분류'], axis=1).columns
  for each_col in tqdm(col_list):
    Q_975 = np.quantile(df[each_col], 0.975)
    Q_025 = np.quantile(df[each_col],0.025)

    df.loc[( Q_025 > df[each_col]) | (df[each_col] > Q_975), each_col] = np.nan

  return df

df_07_after = outlier_replace_nan_v2(df_07)
df_07['NaN_cnt'] = df_07.isnull().sum(1)
```
