```python
import yfinance as yf
import snowball as sb

data = yf.download('069500.KS 114820.KS', period='MAX')['Adj Close']

portfolio = sb.run_backtest(prices=data, 
                            schedule='EOY-5', 
                            rule={'069500.KS': 0.6, '114820.KS': 0.4},
                            start='2012-12-01')
portfolio.report()
```
