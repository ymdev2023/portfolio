```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
from collections import Counter
from wordcloud import WordCloud


# Load the data
data = pd.read_csv('dataset.csv', sep="\t")  # Adjust the dataset file path
data
```

```python
# Preprocess the text
texts = data['content'].values
labels = data['label'].values
texts
```

```python
import re

# Define a function to remove special characters and numbers
def remove_special_characters(text):
    # 특수문자 및 숫자 제거
    text = re.sub(r'[^A-Za-z가-힣\s]', '', text)
    return text

# 모든 문장에 대해 특수문자를 제거
texts = [remove_special_characters(text) for text in texts]
texts
```

```python
# 데이터셋을 문장으로 추출 (label이 1인 경우만)
labeled_texts = data[data['label'] == 1]['content'].values


# 텍스트를 모두 이어붙여서 하나의 문자열로 만듭니다.
all_text = ' '.join(labeled_texts)

# 모든 텍스트를 공백을 기준으로 단어로 분할하고 빈도를 계산합니다.
word_counts = Counter(all_text.split())

# 가장 많이 나오는 단어 100개를 추출합니다.
top_words = word_counts.most_common(100)

# 한글 폰트 경로 지정 (나눔 폰트 경로에 맞게 수정)
font_path = '/Library/Fonts/AppleGothic.ttf' 

# 워드 클라우드 생성
wordcloud = WordCloud(
    font_path=font_path,  # 한글 폰트 경로 지정
    width=800,
    height=400,
    background_color='white'
).generate_from_frequencies(dict(top_words))

# 워드 클라우드를 표시합니다.
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Tokenizer를 이용한 텍스트 전처리
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

# 텍스트를 정수 시퀀스로 변환
sequences = tokenizer.texts_to_sequences(texts)

# 시퀀스를 일정한 길이로 패딩
max_sequence_length = 100  # 문장 최대 길이에 맞게 조절
sequences = pad_sequences(sequences, maxlen=max_sequence_length)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(sequences, labels, test_size=0.2, random_state=42)
```

```python
y_train = y_train.reshape(-1, max_sequence_length, 1)
y_test = y_test.reshape(-1, max_sequence_length, 1)

import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Conv1D, GlobalMaxPooling1D, Dense, Flatten, Attention

# 모델 정의
input_layer = Input(shape=(max_sequence_length,))
embedding_layer = Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=128)(input_layer)
conv1 = Conv1D(64, 3, activation='relu')(embedding_layer)
conv2 = Conv1D(128, 3, activation='relu')(conv1)
pooling = GlobalMaxPooling1D()(conv2)

# Attention Mechanism 추가
attention = Attention()([embedding_layer, pooling])

# Dense 레이어를 통해 감정 분류 수행
dense1 = Dense(128, activation='relu')(attention)
output_layer = Dense(1, activation='sigmoid')(dense1)

model = tf.keras.Model(inputs=input_layer, outputs=output_layer)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=64, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

VDCNN (Very Deep Convolutional Networks for Text Classification)은 텍스트 분류에 효과적인 아키텍처로, 텍스트 데이터를 학습하기 위한 깊은 합성곱 신경망(Convolutional Neural Network, CNN) 기반의 모델입니다. VDCNN은 각 레이어에서 많은 필터와 깊은 네트워크를 사용하여 텍스트의 다양한 특징을 추출하고 감지합니다. 아래에 코드를 짠 이유와 간략한 설명을 제공하겠습니다.

1. **Embedding Layer**: 코드에서 사용된 `Embedding` 레이어는 단어 임베딩을 학습하기 위한 것입니다. 이 레이어는 각 단어를 밀집 벡터로 매핑하고, 단어의 의미와 관계를 포착합니다. 이것은 모델이 텍스트 데이터를 이해하는 데 중요합니다.

2. **Convolutional Layers**: 코드에서 사용된 `Conv1D` 레이어는 1차원 컨볼루션 레이어입니다. 이 레이어는 텍스트에서 패턴과 특징을 감지하는 역할을 합니다. 여러 레이어를 사용함으로써 모델은 점차 더 복잡한 특징을 학습할 수 있습니다.

3. **Pooling Layer**: `GlobalMaxPooling1D` 레이어는 각 합성곱 레이어 다음에 사용됩니다. 이 레이어는 각 레이어에서 가장 중요한 특징만 추출하여 데이터를 축소합니다. 이것은 모델의 파라미터 수를 줄이고, 과적합을 방지하며, 텍스트의 중요한 특징을 강조하는 데 도움을 줍니다.

4. **Dense Layers**: 코드에서 사용된 `Dense` 레이어는 완전 연결된 레이어로, 이 레이어를 통해 특징을 분류합니다. 여기서는 이진 분류를 수행하므로 하나의 뉴런으로 된 출력 레이어를 사용하며, `sigmoid` 활성화 함수를 사용하여 확률을 출력합니다.

5. **모델 컴파일 및 학습**: 모델은 `compile` 메서드를 사용하여 손실 함수와 최적화 알고리즘을 설정하고, `fit` 메서드를 사용하여 학습 데이터에 대해 모델을 학습시킵니다. 이 코드에서는 이진 분류를 수행하기 때문에 `binary_crossentropy` 손실 함수를 사용하였고, `adam` 최적화 알고리즘을 선택하였습니다.

VDCNN은 텍스트 분류에 강력한 성능을 보이며, 깊고 복잡한 텍스트 특징을 감지하기 위한 레이어 구성을 가지고 있습니다. 이러한 아키텍처는 텍스트 데이터의 특징을 효과적으로 추출하고 부정적인 레이블을 감지하는 데 도움을 줄 것으로 기대됩니다. 코드는 데이터셋 및 모델의 특수한 요구에 따라 조정될 수 있으며, 하이퍼파라미터를 튜닝하여 모델의 성능을 향상시킬 수 있습니다.
