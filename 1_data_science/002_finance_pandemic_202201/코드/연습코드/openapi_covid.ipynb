### 라이브러리 import 및 url 불러오기

```python
import requests 
import pprint
import json
import pandas as pd
from pandas.io.json import json_normalize
import xmltodict

api_key = "개인 서비스 키" # 개별 입력
api_key_decode = requests.utils.unquote(api_key) 

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson'
params ={'serviceKey' : api_key_decode, 'pageNo' : '1', 'numOfRows' : '10', 'startCreateDt' : '20201231', 'endCreateDt' : '20211231' }

response = requests.get(url, params=params)
print(response.content)
```

```python
contents = response.text
contents
```

```python
# 예쁘게 보기
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

### json으로 변환하기

```python
# xml을 dictionary 형태로 전환
response_dict= xmltodict.parse(contents)
response_dict
```

```python
response_j = json.dumps(response_dict)
json_ob = json.loads(response_j)

print(json_ob)
print(type(json_ob)) #json타입 확인
```

### 필요한 내용만 남기기

```python
body = json_ob['response']['body']['items']['item']
print(body)
```

### DataFrame 변환

```python
df = json_normalize(body)
df
```

### 번외
- 일일 확진자 컬럼 생성하기
- 파일 저장

```python
df = df[['stateDt', 'decideCnt']] # statDt: 기준일, decideCnt: 확진자 수
df
```

```python
df.info() # object형식임을 알 수 있음
```

```python
df['stateDt'] = pd.to_datetime(df['stateDt']) # 날짜를 datetime 형식으로 
df['decideCnt'] = df['decideCnt'].astype(int) # 확진자 수를 정수로 변환
df
```

```python
df.info()
```

```python
df['decide'] = 0 # 일일확진자 수를 'decide' 컬럼으로 만듦

for i in range(len(df)-1):
    df['decide'][i] = df['decideCnt'][i] - df['decideCnt'][i + 1]
```

```python
df
```

```python
df = df.drop(index=365) # 2020-12-31에 해당하는 마지막 행을 삭제
```

```python
df.rename(columns = {'stateDt' : 'date'}, inplace = True) # 열 이름을 바꾸고
df = df[['date', 'decide']] # 필요한 열만 남긴다
df
```

```python
df.to_csv('dataset/covid.csv') # 파일 저장
```
