### Merging the stock data and news info

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia=SentimentIntensityAnalyzer()
```

```python
df_news = pd.read_json('headlines_june.json',convert_dates=['Date'])
df_news = pd.DataFrame.from_records(df_news['news'])

df_news.head()
```

```python
df_news = pd.read_csv('sent_score_new_en.csv', encoding='utf-8')
# df_news = pd.DataFrame.from_records(df_news['news'])

df_news.head()
```

```python
df_news['Date'] = pd.to_datetime(df_news['Date'])
df_news['Date'] = df_news['Date'].dt.strftime('%d-%b-%Y')

df_bank = pd.read_csv('MCB_01012010_06122018.csv')
df_bank.head()
```

```python
df_news1 = pd.read_pickle("new.pkl")
df_news1.head()
```

```python
df_news.head()
```

```python
df_news = df_news.astype(object)
df_news1
```

```python
df_news1["Date"] = df_news1["Date"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df_news1["Date"] = df_news1["Date"].astype(int)
df_news1["Date"]
```

```python
df_news["Date"] = df_news["Date"].astype(int)
```

```python
df_merged = df_news1.merge(df_news,on='Date', how='outer')
df_merged
```

```python
df_merged.tail()
```

```python
df_merged.shape
```

```python
df_merged=df_merged[['Date','Open','High','Low','Close','Volume','title']]
df_merged.head()
```

```python
# df_merged['Open'] = df_merged['Open'].astype(int)
# df_merged['High'] = df_merged['High'].astype(int)
# df_merged['Low'] = df_merged['Low'].astype(int)
# df_merged['Close'] = df_merged['Close'].astype(int)
```

```python
df_merged['compound']=''
df_merged['neg']=''
df_merged['neu']=''
df_merged['pos']=''

df_merged.head()
```

```python
for index,sentence in enumerate(df_merged['title']):
    ps=sia.polarity_scores(sentence)
    df_merged['compound'][index]=ps['compound']
    df_merged['neg'][index]=ps['neg']
    df_merged['neu'][index]=ps['neu']
    df_merged['pos'][index]=ps['pos']
#   break
```

```python
final_df = df_merged[['Date','compound','neg','neu','pos','Open','High','Low','Close','Volume']]
```

```python
final_df = final_df.dropna(axis=0)
final_df = final_df[:248]
```

```python
final_df.to_csv('final_data_new.csv',index=False)
```
